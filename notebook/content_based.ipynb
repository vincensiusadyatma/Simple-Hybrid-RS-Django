```python
import pandas as pd
```

```python
df = pd.read_csv("../dataset/Hotels-Jakarta.csv")

df
```

```python
df.info()
```

```python
# === 2. Preprocessing ===
df['review_score_text'] = df['review_score_text'].str.replace(',', '').str.replace(' reviews', '').astype(int)
df['hotel_price'] = df['hotel_price'].str.replace('Rp', '').str.replace('.', '').str.replace(',', '').astype(int)
```

```python
# One-hot encode review_score_title
df = pd.get_dummies(df, columns=['review_score_title'])
```

```python
df
```

```python
# === 3. Fitur yang digunakan untuk filtering ===
features = ['review_score','review_score_text', 'hotel_price'] + \
           [col for col in df.columns if col.startswith('review_score_title_')]
```

```python
df.info()
```

```python
df[features]
```

```python
# Normalisasi fitur
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])

# Hitung cosine similarity antar hotel
similarity_matrix = cosine_similarity(scaled_features)
```

```python
# Pilih index hotel sebagai acuan, misalnya hotel pada index 0
hotel_index = 0

# Ambil vektor similarity dari hotel tersebut terhadap semua hotel
similarities = similarity_matrix[hotel_index]

# Urutkan dari yang paling mirip ke paling tidak mirip (kecuali hotel sekarang)
similar_hotel_indices = similarities.argsort()[::-1]  # descending
similar_hotel_indices = similar_hotel_indices[similar_hotel_indices != hotel_index]  # hilangkan hotel sekaranag

# Tampilkan top N hotel yang mirip
top_n = 5
for i in range(top_n):
    idx = similar_hotel_indices[i]
    print(f"{i+1}. {df.loc[idx, 'hotel_name']} (Similarity: {similarities[idx]:.4f})")
```

```python
# === 4. Fungsi Rekomendasi Hotel ===
def recommend_hotel(hotel_index, top_n=3):
    similarity_scores = list(enumerate(similarity_matrix[hotel_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similar_indices = [i[0] for i in similarity_scores[1:top_n+1]]
    return df.iloc[similar_indices][['hotel_name', 'review_score', 'hotel_price']]

# === 5. Contoh: Rekomendasi hotel mirip dengan indeks 0 ===
print("Hotel yang mirip dengan:", df.iloc[0]['hotel_name'])
print(recommend_hotel(0))
```
